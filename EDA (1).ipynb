```python
sns.countplot(x='Gender', hue='Budget', data=df)
plt.title("Count of Gender by Budget")
plt.show()

# Cross-tabulation and percentage calculation
crosstab = pd.crosstab(df['Gender'], df['Budget'])
percentage_crosstab = crosstab.apply(lambda r: round((r / r.sum()) * 100, 1), axis=1)
print(percentage_crosstab)
```

```python
sns.countplot(x='Activity', hue='Overall Rating', data=df)
plt.title("Count of Acivity overall rating")
plt.show()

# Cross-tabulation and percentage calculation
crosstab = pd.crosstab(df['Activity'], df['Overall Rating'])
percentage_crosstab = crosstab.apply(lambda r: round((r / r.sum()) * 100, 1), axis=1)
print(percentage_crosstab)
```

```python
sns.countplot(x='Marital Status', hue='Budget', data=df)
plt.title("Count of Acivity overall rating")
plt.show()

# Cross-tabulation and percentage calculation
percentage_crosstab = crosstab.apply(lambda r: round((r / r.sum()) * 100, 1), axis=1)
print(percentage_crosstab)
```

```python
sns.catplot(data=df,x='Gender',hue='Budget',col='Activity',kind='count',height=4,aspect=1)
plt.suptitle('Gender Distribution across Budget and Activities', y=1.02)
plt.show()

```

```python
sns.catplot(data=df,x='Gender',hue='Cuisines',col='Activity',kind='count',height=4,aspect=1)
plt.suptitle('Gender Distribution across Budget and Activities', y=1.02)
plt.show()
```

```python

# Heatmap for Multivariate Cross-tabulation
crosstab = pd.crosstab(index=[df['Gender'], df['Budget']], columns=df['Activity'])
plt.figure(figsize=(10, 6))
sns.heatmap(crosstab, annot=True, fmt="d", cmap="Blues")
plt.title("Heatmap of Gender and Budget by Activity")
plt.ylabel("Gender and Budget")
plt.xlabel("Activity")
plt.show()
```

```python
# Select only the numerical columns
numerical_columns = ['User ID', 'Area code', 'YOB', 'Budget', 'Food Rating', 'Service Rating', 'Overall Rating']
numerical_data = df[numerical_columns]

# Calculate the correlation matrix
correlation_matrix = numerical_data.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Variables")
plt.show()
```
